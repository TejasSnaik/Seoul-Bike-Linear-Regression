Detecting and removal of Outliers

```python
Q1 = np.percentile(data_re["Rented Bike Count"], 25, interpolation="midpoint")
Q2 = np.percentile(data_re["Rented Bike Count"], 50, interpolation="midpoint")
Q3 = np.percentile(data_re["Rented Bike Count"], 75, interpolation="midpoint")

IQR = Q3-Q1
low_limit = Q1 -1.5 * IQR
up_limit = Q3 + 1.5 * IQR
```

```python
data_re_out = data_re
data_re_out1 = data_re_out[data_re_out["Rented Bike Count"]< up_limit]
data_re_out1 = data_re_out1[data_re_out1["Rented Bike Count"]> low_limit]
data_re_out1
```

```python
Q1 = np.percentile(data_re["Wind speed (m/s)"], 25, interpolation="midpoint")
Q2 = np.percentile(data_re["Wind speed (m/s)"], 50, interpolation="midpoint")
Q3 = np.percentile(data_re["Wind speed (m/s)"], 75, interpolation="midpoint")

IQR = Q3-Q1
lower_limit = Q1 -1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python

data_re_out1 = data_re_out1[data_re_out1["Wind speed (m/s)"]<upper_limit]
data_re_out1 = data_re_out1[data_re_out1["Wind speed (m/s)"]>lower_limit]
```

Setting the Independent Variables responsible for the Output Variable

```python
X = data_re_out1.drop(columns=['Rented Bike Count'])
```

```python
X
```

Setting the Target Variable

```python
y = data_re_out1["Rented Bike Count"]
y
```

Preprocessing the Units of each column data into a common scale

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = MinMaxScaler()
X_scale = scaler.fit_transform(X)
X_scale
```

Splitting the Data into Train and Test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scale, y, test_size=0.2)
```

```python
print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
print("y_train: ", y_train.shape)
print("y_test: ", y_test.shape)
```

Creating an Object of Linear Regression Algorithm

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
```

Training the Linear Regression Algorithm to make predictions 

```python
LR.fit(X_train, y_train)
```

Residual Analysis on model

```python
from sklearn.metrics import mean_squared_error,r2_score
pred_train = LR.predict(X_train)
mse_LR = mean_squared_error(y_train, pred_train)
print("R squared score", r2_score(y_train, pred_train))
```

```python
plt.scatter(x=y_train, y=pred_train)
plt.title("Residual Analysis on train data")
plt.xlabel("y_train")
plt.ylabel("pred_train")
```

As the variance is not constant, so the training data has Heteroscediacity which means the bias is high as training Accuracy is low

```python
pred_y = LR.predict(X_test)
```

Evaluating the Linear Regression Model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print("Accuracy: ", LR.score(X_test, y_test))
print("Mean Absolute Error: ",mean_absolute_error(pred_y , y_test))
print("Mean Squared Error: " , mean_squared_error(pred_y, y_test))
print("R-squared score: ", r2_score(y_test, pred_y))
```

```python
pd.DataFrame({"Predicted": pred_y, "Actual": y_test})
```

```python
plt.scatter(y_test, pred_y)
z=np.polyfit(y_test, pred_y, 1)
p=np.poly1d(z)
plt.plot(y_test, p(y_test), "#F8766D")
```

Regularization

Ridge(L2) Regression

```python
from sklearn.linear_model import Ridge

ridge = Ridge()
ridge
ridge.fit(X_train,y_train)
rd=ridge.predict(X_test)
ridge_score = ridge.score(X_test, y_test)

print(ridge.coef_)
print("R-squared score:", r2_score(rd,y_test))
print("Ridge score",ridge_score)
```

Compairing the predictions made using Ridge

```python
pd.DataFrame({"Predicted": rd, "Actual": y_test})
```

```python
plt.scatter(x=y_test, y=rd)
r=np.polyfit(y_test, rd, 1)
p_r=np.poly1d(r)
plt.plot(y_test, p_r(y_test), "#F8766D")
```

Lasso(L1) Regression

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso
lasso.fit(X_train,y_train)
ls = lasso.predict(X_test)
lasso_score = lasso.score(X_test, y_test)

print(lasso.coef_)
print("R-squared score:", r2_score(ls ,y_test))
print("Lasso score",lasso_score)
```

Compairing prediction made using Lasso

```python
pd.DataFrame({"Predicted": ls, "Actual": y_test})
```

```python
plt.scatter(x=y_test, y=ls)
l=np.polyfit(y_test, ls, 1)
p_L=np.poly1d(l)
plt.plot(y_test, p_L(y_test), "#F8766D")
```
